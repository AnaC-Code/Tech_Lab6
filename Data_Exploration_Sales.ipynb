# Scatter-Plots for the relationships between City, State, Postal_Code, Region, Sales

```python
#loading libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '17072024_sales_data/Clean_Data.csv'
data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
# Scatter plot for Sales vs. Region 
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Region', y='Sales', hue='Region', palette='viridis', s=100)
plt.title('Sales Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.legend(title='Region')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Scatter plot for Sales vs. State
plt.figure(figsize=(12, 8))
sns.scatterplot(data=data, x='State', y='Sales', hue='Region', palette='viridis', s=100)
plt.title('Sales Distribution by State')
plt.xlabel('State')
plt.ylabel('Sales')
plt.legend(title='Region')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
# Scatter plot for Sales vs. City
plt.figure(figsize=(14, 10))
sns.scatterplot(data=data, x='City', y='Sales', hue='Region', palette='viridis', s=100)
plt.title('Sales Distribution by City')
plt.xlabel('City')
plt.ylabel('Sales')
plt.legend(title='Region')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
#number of cities in the data list
unique_cities = data['City'].nunique() # .nunique() counts the number of different Values 
unique_cities
```

```python
# Get the first 25 values
data_subset = data.head(25)

# Scatter plot for Sales vs. City (first 25 values)
plt.figure(figsize=(14, 10))
sns.scatterplot(data=data_subset, x='City', y='Sales', hue='Region', palette='viridis', s=100)
plt.title('Sales Distribution by City (First 25 Values)')
plt.xlabel('City')
plt.ylabel('Sales')
plt.legend(title='Region')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
# Aggregate sales by city and get the top 25 cities by sales
top_cities = data.groupby('City')['Sales'].sum().nlargest(25).index
top_cities_data = data[data['City'].isin(top_cities)]

# Scatter plot for Sales vs. City (top 25 cities by sales)
plt.figure(figsize=(14, 10))
sns.scatterplot(data=top_cities_data, x='City', y='Sales', hue='Region', palette='viridis', s=100)
plt.title('Sales Distribution by City (Top 25 Cities by Sales)')
plt.xlabel('City')
plt.ylabel('Sales')
plt.legend(title='Region')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
# Create scatter plot for Sales vs. Postal_Code (all values)
plt.figure(figsize=(14, 10))
sns.scatterplot(data=data, x='Postal_Code', y='Sales', hue='Region', palette='viridis', s=100)
plt.title('Sales Distribution by Postal Code (All Values)')
plt.xlabel('Postal Code')
plt.ylabel('Sales')
plt.legend(title='Region')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
# Convert Postal_Code to string for better categorical plotting
data['Postal_Code'] = data['Postal_Code'].astype(str)

# Aggregate sales by postal code and get the top 25 postal codes by sales
top_postal_codes = data.groupby('Postal_Code')['Sales'].sum().nlargest(25).index
top_postal_codes_data = data[data['Postal_Code'].isin(top_postal_codes)]

# Create scatter plot for Sales vs. Postal_Code (top 25 postal codes by sales)
plt.figure(figsize=(14, 10))
sns.scatterplot(data=top_postal_codes_data, x='Postal_Code', y='Sales', hue='Region', palette='viridis', s=100)
plt.title('Sales Distribution by Postal Code (Top 25 Postal Codes by Sales)')
plt.xlabel('Postal Code')
plt.ylabel('Sales')
plt.legend(title='Region')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```
