```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = r'C:\Users\User\Documents\GitHub\Tech_Lab6\17072024_sales_data\Clean_Data.csv'
try:
    df = pd.read_csv(file_path, encoding='ISO-8859-1')
    print("File loaded successfully.")
except FileNotFoundError:
    print(f"File not found: {file_path}")
    raise
except Exception as e:
    print(f"Error loading file: {e}")
    raise

# Ensure the 'City' column exists in the DataFrame
if 'City' not in df.columns:
    raise ValueError("Column 'City' not found in the data.")

# Identify the top 7 best-sold cities
top_cities = df['City'].value_counts().nlargest(7)

# Plot bar chart for the top 7 cities
plt.figure(figsize=(10, 6))
top_cities.plot(kind='bar')
plt.title('Top 7 Best-Sold Cities')
plt.xlabel('City')
plt.ylabel('Number of Sales')
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
